## Load data

```python
gmd = pd.read_csv('data/GMD.csv')
print(gmd.columns)
gmd = gmd.pivot(index='year', columns='ISO3')
```

```python
gdp_per = gmd['rGDP_USD'] / gmd['pop']
gdp_per[['USA', 'CHN']].plot(logy=True);
```
